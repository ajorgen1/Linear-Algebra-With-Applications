## 7.1 Matrix Algebra

A <u>matrix</u> is the result of combining similar row or column vectors into a single mathematical object.

A <u>tensor</u> is an algebraic object that describes a multilinear relationship between sets of algebraic objects related to a vector space.

For computational purposes we can consider a matrix to be a 2-dimensional (Python "dimension," not math) tensor.

In mathematics terminology this is an *m* by *n* dimensional matrix:

\begin{align} A = \begin{bmatrix}
    a_{11} & a_{12} & \dots \\
    \vdots & \ddots & \\
    a_{m1} &        & a_{mn}
    \end{bmatrix}
\end{align}

### Matrix-Matrix Multiplication.

Suppose $A$ is $m \times n$ and $B$ is $k \times p$. "Standard" matrix multiplication $AB$ is only defined when $n = k$ and the resulting matrix is of size $m \times p$. The most common way to find the entries of new matrix $AB$ is to find the dot products of the rows of $A$ and the columns of $B$.

The <u>transpose</u> of a matrix $A$ is denoted $A^T$, and can be interpreted as making the columns of $A$ the rows of $A^T$, or vice-versa.

## 7.2 Application: Linear Dynamical Systems

Suppose that a vector $\vec{x}_t = <u_1, u_2, ... u_n>$ gives the "state" of the system at time $t$. Linear Dynamical Systems (often called "Markov Processes") follow an *update equation*: $\vec{x}_{t+1} = A\vec{x}_t$ for $t = 0, 1, 2, ...$

We can use either matrix algebra or simulation to find the state of the system at time $t$ in the future.

* 5% of susceptible get disease, 95% remain susceptible

* of those infected: 1% die, 10% recover/immune, 4% recover/susceptible, rest remain infected.
