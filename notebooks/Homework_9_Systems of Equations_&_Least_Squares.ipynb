## 9.5 Model Selection via Forward Selection

In the previous sections we fitted a model using all three features. But is the 3-feature model the *best* model? Adding more features always reduces RMSE on the data used to fit the model — but that does not mean more features is always better. A simpler model with fewer parameters is less likely to "memorize" the quirks of the training data and tends to **generalize** better to new cars the model has never seen. This trade-off between fit quality and model complexity is called the **bias-variance trade-off**, and it is one of the central ideas in machine learning.

A practical strategy for navigating this trade-off is **forward selection**: start with no features and greedily add the one feature at a time that produces the largest drop in RMSE, stopping when you have reached your desired model size. Here we use forward selection to find the best 2-feature model.

The three candidate 2-feature models are:

| Model | Features |
|:-----:|:---------|
| A | Age + Mileage |
| B | Age + Engine size |
| C | Mileage + Engine size |

```python
# Run this cell without modification
def fit_and_evaluate(feature_matrix, target_vector):
    """Fit a least-squares model and return (x_hat, b_hat, rmse)."""
    x_hat = np.linalg.inv(feature_matrix.T @ feature_matrix) @ feature_matrix.T @ target_vector
    b_hat = feature_matrix @ x_hat
    eps   = target_vector - b_hat
    rmse  = np.linalg.norm(eps) / np.sqrt(len(target_vector))
    return x_hat, b_hat, rmse
```
