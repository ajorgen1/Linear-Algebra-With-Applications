### Why Include a $\beta_0$ Intercept?

The key difference between the least squares model you built in [Homework 9](https://colab.research.google.com/drive/1nvms48iVwGSYBuFRvMKQ0VMLQeFM4QNs?usp=sharing) and this linear regression model is this one includes an intercept $\beta_0$. Why include an intercept? Remember that the model's goal is to make predictions ($\hat{y}$) that are as close to the actual $y$ values as possible. Including an intercept gives more mathematical flexibility for the model to achieve that goal. From the perspective of fitting a line to the data, a model without a $\beta_0$ will be forced to go through the origin. It is often the case that a better fit can be achieved by allowing the line to cross the y-axis somewhere other than 0.

There are, of course, trade-offs to be considered. In general, a simpler model is preferred to a more complicated one due to the fact that simpler models




1. Tend to perform better on data that the model wasn't trained on.
2. Are easier to interpret.
3. Require less computational power (this used to be a much bigger deal than it is now).

  "Tend to perform better on data that the model wasn't trained on" is the most important reason for keeping models simple. A more complicated model that has extra parameters (which are the $\beta_i$s in a linear regression model) will tend to suffer from "over-fitting." Over-fitting happens when the model *learns the patterns and quirks of the data set it was trained on* as compared to *learning the patterns that generalize to making predictions on novel data*.

So the question we should ask is: "Is it worth it to include an intercept $\beta_0$ in our regression model?" The answer is, as always, dependent on the specifics of the use-case, but is generally "yes."
