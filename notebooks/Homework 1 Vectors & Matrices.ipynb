**Summation notation** gives us a compact way to write the sum of the components of a vector. For a vector $\mathbf{x} \in \mathbb{R}^n$, the sum of its components is:

$$\sum_{i=1}^{n} x_i = x_1 + x_2 + \cdots + x_n.$$

The $\Sigma$ (capital sigma, a Greek "S") means "add up," the $i = 1$ at the bottom tells you where to start, and the $n$ at the top tells you where to stop. The index $i$ steps through each value from $1$ to $n$, adding $x_i$ each time.

**Example.** For $ \mathbf{a} = \begin{bmatrix} 4 \\ -1 \\ 2 \end{bmatrix}$:

$$\sum_{i=1}^{3} a_i = a_1 + a_2 + a_3 = 4 + (-1) + 2 = 5$$
