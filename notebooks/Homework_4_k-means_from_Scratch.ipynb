````python
# Helper functions for plotting — run without modification

def plot_iteration(X_small, group_labels, c_0, c_1, c_2, title):
    colors = ['steelblue', 'tomato', 'seagreen']
    fig, ax = plt.subplots(figsize=(5, 5))
    for i, xi in enumerate(X_small):
        g = group_labels[i]
        ax.scatter(xi[0], xi[1], color=colors[g],
                   edgecolors='white', linewidth=0.5, s=80, zorder=3)
        ax.annotate(f'$\\mathbf{{x}}_{{{i}}}$', xy=(xi[0], xi[1]),
                    xytext=(5, 5), textcoords='offset points', fontsize=9)
    ax.scatter(c_0[0], c_0[1], color=colors[0], marker='*', s=350,
               edgecolors='black', linewidth=0.8, zorder=4,
               label=f'$\\mathbf{{c}}_0$ = {np.round(c_0, 2)}')
    ax.scatter(c_1[0], c_1[1], color=colors[1], marker='*', s=350,
               edgecolors='black', linewidth=0.8, zorder=4,
               label=f'$\\mathbf{{c}}_1$ = {np.round(c_1, 2)}')
    ax.scatter(c_2[0], c_2[1], color=colors[2], marker='*', s=350,
               edgecolors='black', linewidth=0.8, zorder=4,
               label=f'$\\mathbf{{c}}_2$ = {np.round(c_2, 2)}')
    ax.set_xlabel('$A$')
    ax.set_ylabel('$B$')
    ax.set_title(title)
    ax.legend(fontsize=9)
    ax.grid(True, linestyle='--', alpha=0.4)
    plt.tight_layout()
    plt.show()
````
