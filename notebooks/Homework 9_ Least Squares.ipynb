## 8.3 Inverse Matrices

**Exercise 8**. The system $A\vec{x} = \vec{b}$:

\begin{align} \begin{bmatrix}
    -1 & 4 & 1 \\
    3 & 0 & 2 \\
    1 & -1 & 0\\
    \end{bmatrix}
    \begin{bmatrix}
    x_1   \\

    x_3  \\
    \end{bmatrix} =
    \begin{bmatrix}
    -3  \\

    2 \\
    \end{bmatrix}
\end{align}

is guaranteed to have an inverse $A^{-1}$ because # rows = # cols = rank(A) = \_\_\_.

````python
import numpy as np

x = np.linalg.inv(np.array([[-1, 4, 1], [3, 0, 2], [1, -1, 0]]))@np.array([[-3], [7], [2]])
x
````
