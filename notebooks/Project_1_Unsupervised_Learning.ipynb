### Why exploration comes first

Choosing a dataset is not like choosing a homework problem where the answer is guaranteed to exist. Real data is unpredictable, and a dataset that sounds interesting can turn out to be flat, poorly structured, or just not well-suited to k-means. The professional habit is to explore several candidates before committing to one.

For each candidate dataset you consider, you should at minimum do the following in your scratch notebook:

- Load it and check its shape, column names, and data types.
- Compute column statistics — means, standard deviations, norms. Are the features on wildly different scales? Are any features nearly constant, or obviously redundant?
- Make a scatter plot or two. Does the data show any visual structure — clusters, separations, gradients — or does everything collapse into one blob?
- Run a quick k-means with a few values of $k$ and look at the resulting elbow plot. Does $J^{\text{clust}}$ drop meaningfully as you add clusters, or does it decrease almost linearly (a signal that there may not be much cluster structure to find)?
- Apply PCA and check how much variance the first two components explain. If PC1 + PC2 together explain only 20% of the variance, your final 2D cluster visualization will be nearly uninterpretable.

None of this pre-work needs to be polished. The goal is just to get enough of a signal to know whether a dataset is worth committing to. When you find one where the scatter plots suggest structure, the elbow plot has a visible bend, and PCA explains a meaningful fraction of the variance, that is your dataset.

### What makes a dataset suitable

- At least 4 numerical features.
- At least 100 rows.
- Real, meaningful features — not arbitrary ID numbers or raw timestamps.
- Some domain reason to believe groups might exist. This could be prior knowledge, a visual hint from a scatter plot, or just a reasonable hypothesis worth testing.

[Kaggle](https://www.kaggle.com/datasets), the [UCI Machine Learning Repository](https://archive.ics.uci.edu/), and [Google Dataset Search](https://datasetsearch.research.google.com/) are all good starting points. Any real tabular dataset meeting the criteria above is acceptable.

Before you commit to a dataset, ask yourself a sharper question than "does k-means run on this?" Ask: **what would you actually learn from grouping these rows into clusters?** K-means is not an end in itself — it is a tool for discovering structure that is meaningful to someone.

Think about who would care about these results. If you are working with health data, are there patient subgroups that might respond differently to treatment? If you are working with economic data, do the clusters correspond to different spending behaviors or risk profiles? If you are working with environmental data, do the clusters reveal geographic or seasonal patterns? You do not need a definitive answer before you start — but you should have a hypothesis worth testing.
