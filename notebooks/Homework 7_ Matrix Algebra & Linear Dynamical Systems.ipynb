## 7.1 Linear Dynamical Systems in Pharmaco-Kinetics

The following model is adapted from *Introduction to Applied Linear Algebra* by Boyd and Vandenberghe (2018).

A *compartmental system* is a model used to describe the movement of some material over time among a set of *n* compartments of a system, and the outside world. It is widely used in *pharmaco-kinetics*, the study of how the concentration of a drug varies over time in the body. In this application, the material is a drug, and the compartments are the bloodstream, lungs, heart, liver, kidneys, and so on. Compartmental systems are special cases of linear dynamical systems.

In the following exercises we will consider a very simple compartmental system with 3 compartments. We let $\vec{x}_t$ denote the amount of material (drug) in the system at time period $t$. Between period $t$ and $t+1$, the material moves as follows.



* 10% of the material in compartment 1 moves to compartment 2.
* 5% of the material in compartment 2 moves to compartment 3.
* 5% of the material in compartment 3 moves to compartment 1.
* 5% of the material in compartment 3 is eliminated.

Express this compartmental system as a linear dynamical system, $\vec{x}_{t+1} = A\vec{x}_t$.

### Safety Mechanisms in Modern Medicine.

It used to be that medical professionals had to calculate proper dosages of medications using charts or common rules-of-thumb. These days with access to computers and data hospitals and clinics are increasingly using software to recommend medications, calculate suggested dosages, and catch medication administration errors before they happen.

Let's think about how this might work in the context of our 3-Organ Markov problem. Suppose that medical staff decide to inject 10 milligrams of medication into Organ 1. There is some risk involved in this because if at any point in time there is too much medication in Organ 3 there will be toxic side effects.

Additionally, research has shown that Organ 3's exposure to amount of medication over time (measured in $milligrams \cdot hours$) also matters; if there is prolonged exposure this may also cause toxcicity. For those of you who have taken Calculus 2, if $f(t)$ is the amount of medication in Organ 3 at time t, then $mg \cdot hrs$ can be represented by the improper integral:

\begin{align}
\int_{0}^{\infty} f(t)dt,
\end{align}

which we can numerically estimate using the ```x_t[2]``` values generated by our Markov equation and ```np.trapz```, which estimates the area under the "Mg. of Medication in Organ 3" curve.


The following function is an example of how a safety check can be programmed and used as software. Run the code cell without modification.


````python
def safety_check(A, x_0):
  y = np.array([x_0[2]])
  n = 1000

  for i in range(1, n+1):
    x_i = np.linalg.matrix_power(A,i)@x_0 # Note: This line assumes that you have already created and stored matrix A and vector x_0!
    y = np.append(y,x_i[2])

  if np.max(y) > 3:
    return "Warning: Initial dose results in Organ 3 exceeding 3 milligrams."
  elif np.trapz(y) > 150:
    return "Warning: Initial dose results in Organ 3 exceeding 150 mg*hr."
  else:
    return "Initial dose is safe."
````

## 7.3 Epidemic Dynamics Model with Vaccination and Immigration

Let's think about how we would extend the model in Exercise 10 to include immigration. In particular, consider the following immigration policy:

* From $t=0$ and all time-periods onward $0.5\%$ immigration is allowed into the population per time period.

* No one is allowed to immigrate into the population if they are infected or deceased.

* Of the $0.5\%$ who immigrate at each time period, a random fraction are susceptible, the rest are recovered/immune. At each time period the fraction that are susceptible is drawn from a $Uniform \sim [0, 1]$ distribution.

This modification of the model results in a more complicated Markov update equation. Let $\vec{v}_t$ be the immigration vector at time $t$. The update equation now becomes

\begin{align}
\vec{x}_{t+1} = A\vec{x}_t + \vec{v}_t.
\end{align}
