## 3.1 The Dot Product

How should we define "vector-vector multiplication"? So far we have done vector-scalar multiplication: $\lambda\vec{v}$. We know that this kind of multiplication changes the length and (possibly) the direction of $\vec{v}$.

One of the most useful ways to multiply vectors is the <u>dot product</u> (also called the "scalar" or "inner" product).

**Definition**. For two vectors $\vec{u}$ and $\vec{v}$:

$\vec{u} \cdot \vec{v} = \Sigma_{i=1}^n u_iv_i$.

### Matrix Multiplication.

**Definition**. A <u>matrix</u> is a mathematical object that consists of an *ordered* collection of vectors. A general *m* by *n* dimensional matrix is:

\begin{align} A = \begin{bmatrix}
    a_{11} & a_{12} & \dots \\
    \vdots & \ddots & \\
    a_{m1} &        & a_{mn}
    \end{bmatrix} .
\end{align}

We will study matrices and their properties in the second half of this course. One of the key concepts that we will learn is that matrix-matrix multiplication $AB$ (think matrix $A$ times matrix $B$, in that order) is only defined if $A$ has the same number of columns as the number of rows in $B$. Furthermore, $AB$ will be a matrix that has the same number of rows as $A$ and same number of columns as $B$.

Why is this coming up now? It can be helpful to think of vectors as "thin" matrices.

### Properties of the Dot Product.

a) $\vec{u} \cdot \vec{v} = 0$ if and only if $\vec{u}$ and $\vec{v}$ are "orthogonal."

b) If $\vec{u}$ and $\vec{v}$ point in the same direction then $\vec{u} \cdot \vec{v} = ||\vec{u}||||\vec{v}||$.

c) If $\vec{u}$ and $\vec{v}$ point in the opposite direction then $\vec{u} \cdot \vec{v} = -||\vec{u}||||\vec{v}||$.
