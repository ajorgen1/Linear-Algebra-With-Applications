### Standard Position vs. Free Vectors

By convention we draw vectors with their *tail* at the origin — this is called **standard position**. But a vector is defined entirely by its **magnitude** (length) and **direction**, not by where it starts. The same vector $\mathbf{x} = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$ can be drawn starting from any point in the plane — it is still the same vector.

The plot below shows $\mathbf{x}$ drawn from three different starting points — including one in the third quadrant. All three arrows are identical in length and direction — they are all $\mathbf{x}$.

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Same vector x drawn from three different starting points
starts = [(0, 0), (2, -1), (-3, -2)]  # origin, Q4, Q3
labels = ['standard position', 'translated', 'translated']
alphas = [1.0, 0.7, 0.4]

for (sx, sy), label, alpha in zip(starts, labels, alphas):
    ax.quiver(sx, sy, x[0], x[1], angles='xy', scale_units='xy', scale=1,
              color='steelblue', alpha=alpha, label=label)
    # Mark the tail of each arrow
    ax.plot(sx, sy, 'o', color='steelblue', alpha=alpha, markersize=5)

ax.set_xlim(-4, 6)
ax.set_ylim(-3, 5)
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal')
ax.legend(fontsize=10, loc='upper left')
ax.text(5.3, 0.15, r'$x_1$', fontsize=12)
ax.text(0.1, 4.5, r'$x_2$', fontsize=12)
ax.set_title(r'$\mathbf{x}$ drawn from three different starting points — same vector')

plt.show()
```

This will become important when we think about **vector addition** geometrically. The key takeaway: **location does not define a vector. Magnitude and direction do.**

## 2.2 Vector Addition

**Definition.** The sum of two vectors $\mathbf{x}, \mathbf{y} \in \mathbb{R}^n$ is computed component-wise:

$$\mathbf{x} + \mathbf{y} = \begin{bmatrix} x_1 + y_1 \\ x_2 + y_2 \\ \vdots \\ x_n + y_n \end{bmatrix}$$

For our vectors:

$$\mathbf{x} + \mathbf{y} = \begin{bmatrix} 2 \\ 3 \end{bmatrix} + \begin{bmatrix} -1 \\ 2 \end{bmatrix} = \begin{bmatrix} 1 \\ 5 \end{bmatrix}$$

In NumPy, vector addition uses the `+` operator:

```python
z = x + y
print(z)
```

### Geometric Interpretation: The Parallelogram Law

Geometrically, $\mathbf{x} + \mathbf{y}$ is the diagonal of the parallelogram formed by $\mathbf{x}$ and $\mathbf{y}$. Equivalently, place $\mathbf{y}$ at the tip of $\mathbf{x}$ — the sum is the vector from the origin to where you end up.

```python
fig, ax = plt.subplots()

ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='steelblue', label=r'$\mathbf{x}$')
ax.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='tomato', label=r'$\mathbf{y}$')
ax.quiver(x[0], x[1], y[0], y[1], angles='xy', scale_units='xy', scale=1,
          color='tomato', alpha=0.4, linestyle='dashed')
ax.quiver(y[0], y[1], x[0], x[1], angles='xy', scale_units='xy', scale=1,
          color='steelblue', alpha=0.4, linestyle='dashed')
ax.quiver(0, 0, z[0], z[1], angles='xy', scale_units='xy', scale=1, color='seagreen', label=r'$\mathbf{x} + \mathbf{y}$')

ax.set_xlim(-3, 4)
ax.set_ylim(-1, 7)
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal')
ax.legend(fontsize=12)
ax.text(3.3, 0.15, r'$x_1$', fontsize=12)
ax.text(0.1, 6.5, r'$x_2$', fontsize=12)
ax.set_title(r'Vector addition: $\mathbf{x} + \mathbf{y}$')

plt.show()
```

## 2.3 Scalar Multiplication

**Definition.** Multiplying a vector $\mathbf{x} \in \mathbb{R}^n$ by a scalar $c \in \mathbb{R}$ scales each component:

$$c\mathbf{x} = \begin{bmatrix} cx_1 \\ cx_2 \\ \vdots \\ cx_n \end{bmatrix}$$

Geometrically:
- $c > 1$ stretches the vector
- $0 < c < 1$ shrinks the vector
- $c < 0$ reverses the direction
- $c = 0$ produces the zero vector $\mathbf{0}$

In NumPy, scalar multiplication uses the `*` operator:

```python
print(2 * x)   # stretch
print(-1 * x)  # reverse
```

```python
fig, ax = plt.subplots()

ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1,
          color='steelblue', label=r'$\mathbf{x}$')
ax.quiver(0, 0, (2*x)[0], (2*x)[1], angles='xy', scale_units='xy', scale=1,
          color='steelblue', alpha=0.5, label=r'$2\mathbf{x}$')
ax.quiver(0, 0, (-1*x)[0], (-1*x)[1], angles='xy', scale_units='xy', scale=1,
          color='tomato', label=r'$-\mathbf{x}$')

ax.set_xlim(-5, 6)
ax.set_ylim(-4, 7)
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal')
ax.legend(fontsize=12, loc='upper left')  # second quadrant
ax.text(5.3, 0.15, r'$x_1$', fontsize=12)
ax.text(0.1, 6.5, r'$x_2$', fontsize=12)
ax.set_title(r'Scalar multiplication: $\mathbf{x}$, $2\mathbf{x}$, $-\mathbf{x}$')

plt.show()
```

## 2.4 Magnitude and Norm

**Definition.** The **magnitude** (or **norm**) of a vector $\mathbf{x} \in \mathbb{R}^n$ is its length, written $\|\mathbf{x}\|$:

$$\|\mathbf{x}\| = \sqrt{x_1^2 + x_2^2 + \cdots + x_n^2} = \sqrt{\sum_{i=1}^n x_i^2}.$$

This is the Pythagorean theorem extended to $n$ dimensions. For our vector $\mathbf{x} = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$:

$$\|\mathbf{x}\| = \sqrt{2^2 + 3^2} = \sqrt{4 + 9} = \sqrt{13} \approx 3.6056$$

### A note on `np.linalg`

`linalg` is a **subpackage** of NumPy dedicated to linear algebra operations — norms, determinants, eigenvalues, solving systems of equations, and more. The reason it lives separately from the main NumPy namespace is that under the hood it calls **LAPACK** and **BLAS** — highly optimized Fortran and C libraries that have been the industry standard for numerical linear algebra for decades. These are not general-purpose array operations; they are specialized, carefully tuned routines for exactly the kind of math we're doing in this course. You access them with `np.linalg.function_name()`.

```python
norm_x = np.linalg.norm(x)
norm_y = np.linalg.norm(y)

# f-strings (the f before the quote) let you embed variables directly inside a string.
# The variable or expression goes inside curly braces {}.
# :.4f means: format as a float with 4 decimal places.
print(f'||x|| = {norm_x:.4f}')
print(f'||y|| = {norm_y:.4f}')
```

## 2.5 Unit Vectors and Normalization

**Definition.** A **unit vector** is a vector with magnitude 1. Given any nonzero vector $\mathbf{x}$, we can produce a unit vector pointing in the same direction by dividing by its norm:

$$\hat{\mathbf{x}} = \frac{\mathbf{x}}{\|\mathbf{x}\|}$$

This is called **normalizing** $\mathbf{x}$. The hat notation $\hat{\mathbf{x}}$ ("x-hat") is standard for unit vectors — it signals that the vector has been normalized to have magnitude 1.

*Note: don't confuse this hat with the $\hat{y}$ we saw in Lecture 1. There, the hat meant "predicted value." Here, the hat means "unit vector." Same symbol, different context — something you'll encounter often in math.*

Normalizing strips away magnitude (changes it to 1) and keeps only **direction**.

```python
x_hat = x / np.linalg.norm(x)

print(f'x_hat = {x_hat.T}')                        # .T to print as a row for readability
print(f'||x_hat|| = {np.linalg.norm(x_hat):.4f}')  # should be 1.0
```

```python
fig, ax = plt.subplots()

ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1,
          color='steelblue', label=r'$\mathbf{x}$')
ax.quiver(0, 0, x_hat[0], x_hat[1], angles='xy', scale_units='xy', scale=1,
          color='seagreen', label=r'$\hat{\mathbf{x}}$')

ax.set_xlim(-1, 4)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal')
ax.legend(fontsize=12)
ax.text(3.3, 0.15, r'$x_1$', fontsize=12)
ax.text(0.1, 3.5, r'$x_2$', fontsize=12)
ax.set_title(r'$\mathbf{x}$ and its unit vector $\hat{\mathbf{x}}$')

plt.show()
```

### The row × column requirement

The need for `.T` above is worth understanding once and carrying forward. NumPy's `np.dot()` performs **matrix multiplication**, which requires the inner dimensions to match:

$$\underbrace{(1 \times 2)}_{\mathbf{x}^\top} \times \underbrace{(2 \times 1)}_{\mathbf{y}} = \underbrace{(1 \times 1)}_{\text{scalar}}$$

Without the transpose, both `x` and `y` have shape `(2, 1)`, and NumPy raises a `ValueError` because `(2, 1) × (2, 1)` has mismatched inner dimensions (`1` and `2`).

### Geometric Interpretation

There is an equivalent geometric formula for the dot product:

$$\mathbf{x} \cdot \mathbf{y} = \|\mathbf{x}\| \|\mathbf{y}\| \cos\theta$$

where $\theta$ is the angle between $\mathbf{x}$ and $\mathbf{y}$. This connects the algebraic definition to geometry and tells us:

- $\mathbf{x} \cdot \mathbf{y} > 0$ — the vectors point in a *similar* direction ($\theta < 90°$)
- $\mathbf{x} \cdot \mathbf{y} = 0$ — the vectors are **orthogonal**
- $\mathbf{x} \cdot \mathbf{y} < 0$ — the vectors point in *opposite* directions ($\theta > 90°$)

We can rearrange to find the angle between $\mathbf{x}$ and $\mathbf{y}$:

$$\cos\theta = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \|\mathbf{y}\|}$$

```python
cos_theta = np.dot(x.T, y)[0][0] / (np.linalg.norm(x) * np.linalg.norm(y))
theta = np.degrees(np.arccos(cos_theta))

# f-string reminder: {variable:.4f} prints the variable as a float with 4 decimal places
print(f'cos(theta) = {cos_theta:.4f}')
print(f'theta = {theta:.4f} degrees')
```

### Orthogonal vs. 90 Degrees

You may be wondering: why say **orthogonal** instead of just "90 degrees apart"?

In $\mathbb{R}^2$ and $\mathbb{R}^3$, the two are equivalent — orthogonal vectors look perpendicular when you draw them. But as we saw earlier, we can't draw vectors in $\mathbb{R}^4$ or higher. The geometric picture breaks down, and with it the intuitive meaning of "angle."

**Orthogonal** is the algebraic generalization: $\mathbf{x}$ and $\mathbf{y}$ are orthogonal whenever $\mathbf{x} \cdot \mathbf{y} = 0$, regardless of dimension. This definition works in any $\mathbb{R}^n$ — even $\mathbb{R}^{1000}$. The angle formula $\cos\theta = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\|\|\mathbf{y}\|}$ technically still *defines* an angle in higher dimensions, but there is no picture to go with it.

This is a recurring theme in this course: we use low-dimensional pictures to build intuition, then rely on algebra to carry that intuition into dimensions we can't see. **Orthogonal** is the word that travels. We'll use it throughout.
