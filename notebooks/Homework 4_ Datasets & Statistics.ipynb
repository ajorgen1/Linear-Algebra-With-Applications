## 4.2 Correlation and the Iris Dataset

The <u>Pearson Correlation Coefficient</u> measures of how closely two different data vectors (labeled $\vec{x}$ and $\vec{y} \in \mathbb{R}^n$) correspond with each other in a linear relationship.

\begin{align}
r = \frac{\Sigma_{i=1}^n(x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\Sigma_{i=1}^n(x_i-\bar{x})^2}\sqrt{\Sigma_{i=1}^n(y_i-\bar{y})^2}} = (\vec{z}_x \cdot \vec{z}_y)/n.
\end{align}

Remember our interpretation of the dot product?

* A positive dot product indicates that the vectors are pointed in a similar direction.

* A 0 dot product indicates the vectors are "orthogonal."

* A negative dot product indicates that the vectors are pointed in opposite directions.

The correlation coefficent $r$ is simply a scaled version of the dot product: its formula is fairly compicated but is equivalent to finding the average of the dot product of the standardized vectors: $(\vec{z}_x\cdot \vec{z}_y)/n$.

By scaling the vectors in this way ensures that $r$ will always be in the interval $[-1, 1]$.

* $r \approx -1$ indicates *strong negative correlation*: when the entries in $\vec{x}$ are large positive the entries in $\vec{y}$ tend to be large negative.

* $r \approx 0$ indicates *little or no correlation*: there's no predictive relationship between the entries of $\vec{x}$ and $\vec{y}$.

* $r \approx 1$ indicates *strong positive correlation*:  when the entries in $\vec{x}$ are large positive the entries in $\vec{y}$ tend to be large positive.

**The Takeaway**: $r$ is a scaled version of the dot product of two vectors and it can be interpreted similarly to how we would interpret the angle between two vectors. As we develop our intuition around vectors and how they relate to each other we will see that $r$ is a measure of how much predictive *information* one vector holds about another.
