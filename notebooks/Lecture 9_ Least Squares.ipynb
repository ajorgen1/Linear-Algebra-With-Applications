## 9.1 Review of Solving System $A\vec{x} = \vec{b}$

**Key Concept**. We can view solving $A\vec{x} = \vec{b}$ for $\vec{x}$ as "find the weights to multiply the columns of $A$ so that the linear combination of these weights times the columns produces $\vec{b}$ as a result."

Suppose that $A$ is $m \times n$, and $A = [\vec{a}_1 | \vec{a}_2 | ... | \vec{a}_n]$ where each $\vec{a}_i$ represent the columns of $A$ and $\vec{x} = <x_1, x_2, ... x_n>$. Then $A\vec{x} = \vec{b}$ is equivalent to

\begin{align}
x_1\vec{a}_1 + x_2\vec{a}_2 + ... + x_n\vec{a}_n = \vec{b}.
\end{align}

### Solutions to Matrix-Vector Equations.

A solution $\vec{x}$ to the "system" $A\vec{x} = \vec{b}$ will fall into one of the three following cases:

1. There are no $x_i$ values that solve the system. It is impossible to get vector $\vec{b}$ from a linear combination of the columns of $A$. This is the "no solution" case.
2. There are only one set of $x_i$ values (equivalently, a unique vector $\vec{x}$) that solve the system. This is called the "one solution" or "unique solution" case.
2. There are many different combinations of $x_i$ values that will make linear combinations that produce vector $\vec{b}$. This is the "infinite number of solutions" case.

### Column Space.

If we consider the columns of a matrix to be individual vectors $\vec{a}_1$, $\vec{a}_2$, ..., $\vec{a}_n$, then we can define the <u>column space</u> of a matrix $A$ as

\begin{align}
C(A) & = \{x_1\vec{a}_1 + x_2\vec{a}_2 + ... + x_n\vec{a}_n : x_i \in \mathbb{R}\}.
\end{align}

"For a given vector $\vec{b}$, do there exist weights $x_1, ..., x_n$ such that I can get $\vec{b}$ via linear combinations of the columns of matrix $A$?" If yes, then $\vec{b}$ is in the column space of $A$.

## 9.2 The Least Squares Solution

Any time we work with real-world data and try to find a solution to $A\vec{x} = \vec{b}$ we will end up in the \_\_\_\_\_\_\_ solution(s) case. (Why?)

**Theorem: Transpose and Rank**. Suppose that matrix $A$ is $m \times n$ and has $m > n = r$. Then $A^TA$ also has rank $n = r$.

**Question**. What size is $A^TA$ and how many linearly independent columns does it have?

Suppose that matrix $A$ is $m \times n$ and $rank(A) = n$. We'd like to us solve the system $A\vec{x}=\vec{b}$; that is: find the compenents of vector $\vec{x}$ that weight the columns of $A$ to make a linear combination of vectors that produces $\vec{b}$ as a result.  

There are two issues with doing this.

**Issue 1: Mathematics**. Column vector $\vec{b}$ isn't actually in $C(A)$. Mathematically this means we're in the "no solution" case. So, what we're really looking for is an appproximate solution: What $\vec{x}$ vector will weight the columns of $A$ to make a linear combination that gets the closest to $\vec{b}$? The first step for finding this vector is to multiply $A\vec{x} = \vec{b}$ from the left by $A^T$:

\begin{align}
A^TA\vec{x} = A^T\vec{b}.
\end{align}

**Implications**. Multiplying by $A^T$ gives us a new system. Matrix "$A$" is now $A^TA$ and vector "$\vec{b}$" is now $A^T\vec{b}$. One way to convince yourself that this is an entirely new system is to think about the dimensions: $A^TA$ is $n \times n$ and $A^T\vec{b}$ is $n \times 1$. The original system $A\vec{x} = \vec{b}$ had a $m \times 1$ vector $\vec{b}$ on the right-hand-side.

**Question**. Assuming $m >> n$ how does original RHS $\vec{b}$ compare to new RHS $A^T\vec{b}$?

**Answer**. $A^T\vec{b}$ is a column vector in a much lower-dimensional space than $\vec{b}$ was. $A^T$ is a matrix that is transforming high-dimensional $\vec{b}$, which is not in $C(A)$ and thus resulting in the no solution case to $A\vec{x} = b$, into a vector $A^T\vec{b}$ that is a representation of $\vec{b}$ in a lower-dimensional space.

**Issue 2: Computation**. If $m>>n$ then a solution for $A\vec{x} = \vec{b}$ is going to be challenging to find with a computer because $A^{-1}$ doesn't exist.  According to the *Transpose and Rank Theorem*  multiplying the system from the left with $A^T$ will fix this.



### The Least Squares Solution to $A\vec{x} = \vec{b}$.

1. Multiply $A\vec{x} =\vec{b}$ from the left by $A^T$. This produces new system $A^TA\vec{x}=A^T\vec{b}$.

2. Multiply from the left by $(A^TA)^{-1}$. This produces system $(A^TA)^{-1}A^TA\vec{x} = (A^TA)^{-1}A^T\vec{b}$.

3. $(A^TA)^{-1}A^TA = I$, which leads to system $I\vec{x} = (A^TA)^{-1}A^T\vec{b}$.

4. $I\vec{x} = \vec{x}$, which leads to the solution vector $\hat{x}= (A^TA)^{-1}A^T\vec{b}$.

If $A\vec{x} = \vec{b}$ has no solution then we can think of $\hat{x} =  (A^TA)^{-1}A^T\vec{b}$ as the "best" approximate solution: it is the vector containing weights that do linear combinations with the columns of $A$ that produces a vector that is the closest to $\vec{b}$ that we can achieve.
