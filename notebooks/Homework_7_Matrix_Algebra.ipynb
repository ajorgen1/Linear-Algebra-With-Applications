## 7.2 A Matrix as a Geometric Transformation

Every $2 \times 2$ matrix $M$ defines a transformation that happens in $\mathbb{R}^2$: it takes each point $\mathbf{v} \in \mathbb{R}^2$ and produces a new point $M\mathbf{v}$. Applying $M$ to every corner of a shape (where the corners can be thought of as points or the tips of vectors in standard position) reveals what the matrix *does* geometrically.

We will use

$$M = \begin{bmatrix} 2 & 0.5 \\ 0 & 1.5 \end{bmatrix}$$

and transform the unit square with corners $(0,0)$, $(1,0)$, $(1,1)$, $(0,1)$.

```python
M = np.array([[2.0, 0.5],
              [0.0, 1.5]])

# Corners of the unit square as individual column vectors
c0 = np.array([0., 0.])
c1 = np.array([1., 0.])
c2 = np.array([1., 1.])
c3 = np.array([0., 1.])
corners = [c0, c1, c2, c3]

# Apply M to each corner individually
transformed_corners = [M @ c for c in corners]

print('Original corners and their transforms:')
for i, (c, t) in enumerate(zip(corners, transformed_corners)):
    print(f'  c{i} = {c}^T  ->  M @ c{i} = {np.round(t, 4)}^T')

# Also build the closed-loop versions for plotting
square      = np.array([c0, c1, c2, c3, c0]).T   # shape (2,5)
transformed = np.array([M @ c for c in [c0,c1,c2,c3,c0]]).T
```

## 7.3 The Identity Matrix

The **identity matrix** $I_n$ is the $n \times n$ matrix with 1s on the main diagonal and 0s everywhere else:

$$I_2 = \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}, \qquad I_3 = \begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix}, \qquad I_n = \begin{bmatrix} 1 & & \\ & \ddots & \\ & & 1 \end{bmatrix}$$

It plays the same role in matrix arithmetic that the number 1 plays in ordinary multiplication: multiplying by $I$ leaves everything unchanged. Formally, for any matrix $A$:

$$I_m A = A = A I_n$$

where we use $I_m$ on the left (matching the number of rows of $A$) and $I_n$ on the right (matching the number of columns of $A$).

**Constructing $I_n$ by hand.** To build $I_n$:
- Start with an $n \times n$ grid of zeros.
- Place a 1 in position $(i, i)$ for each $i = 0, 1, \ldots, n-1$ — the entries where row index equals column index are the **main diagonal**.

## 7.4 Determinants

The **determinant** of a square matrix $A$, written $\det(A)$ or $|A|$, is a single number that captures important structural information about the matrix. As we will see, it tells us whether a matrix can transform space in a reversible way, and it is the key tool for finding eigenvalues.

We build up from $2 \times 2$ to $3 \times 3$, first by hand, then with NumPy.

### The $2 \times 2$ Case

For a $2 \times 2$ matrix, the determinant has a simple formula:

$$\det\begin{bmatrix} a & b \\ c & d \end{bmatrix} = ad - bc$$

This is the product of the main diagonal entries minus the product of the off-diagonal entries.

**Example.** For $M = \begin{bmatrix} 4 & 3 \\ 2 & 1 \end{bmatrix}$:

$$\det(M) = (4)(1) - (3)(2) = 4 - 6 = -2$$

Notice that $S = \begin{bmatrix} 2 & 4 \\ 1 & 2 \end{bmatrix}$ has $\det(S) = 0$. Look at the two columns: column 1 is exactly twice column 0. When one column of a matrix is a scalar multiple of another, the rows carry **redundant information** — its determinant is zero. A matrix with $\det = 0$ is called **singular**. A matrix with $\det \neq 0$ is called **non-singular**.

You can think of a singular matrix as one that "collapses" space — it squashes vectors into a lower-dimensional subspace and loses information. A non-singular matrix keeps all dimensions intact.

### The $3 \times 3$ Case: Cofactor Expansion

For a $3 \times 3$ matrix, we expand along the first row. Each entry in the first row multiplies the determinant of the $2 \times 2$ submatrix formed by deleting that entry's row and column, with alternating signs $+, -, +$:

$$\det\begin{bmatrix} a & b & c \\ d & e & f \\ g & h & i \end{bmatrix} = a\,\det\begin{bmatrix}e&f\\h&i\end{bmatrix} - b\,\det\begin{bmatrix}d&f\\g&i\end{bmatrix} + c\,\det\begin{bmatrix}d&e\\g&h\end{bmatrix}$$

**Example.** Let $P = \begin{bmatrix} 1 & 2 & 3 \\ 0 & 4 & 5 \\ 1 & 0 & 6 \end{bmatrix}$.

Expanding along row 0:

$$\det(P) = 1 \cdot \det\begin{bmatrix}4&5\\0&6\end{bmatrix} - 2 \cdot \det\begin{bmatrix}0&5\\1&6\end{bmatrix} + 3 \cdot \det\begin{bmatrix}0&4\\1&0\end{bmatrix}$$

$$= 1(24 - 0) - 2(0 - 5) + 3(0 - 4) = 24 + 10 - 12 = 22$$

### Determinant as Area Scaling

There is a geometric interpretation of the determinant: $|\det(A)|$ is the factor by which $A$ scales areas. Applying $A$ to the unit square produces a parallelogram whose area equals $|\det(A)|$.

```python
# Run this cell without modification
M_area = np.array([[3., 1.],
                   [0., 2.]])

sq = np.array([[0,1,1,0,0],
               [0,0,1,1,0]], dtype=float)
tr = M_area @ sq

det_M_area = np.linalg.det(M_area)
print(f'det(M_area) = {np.round(det_M_area, 4)}')
print(f'|det(M_area)| = {np.round(abs(det_M_area), 4)}')

fig, axes = plt.subplots(1, 2, figsize=(10, 4.5))

for ax, (pts, color, title) in zip(axes, [
    (sq, 'steelblue', 'Unit square  (area = 1)'),
    (tr, 'tomato',    f'Transformed  (area = |det| = {abs(det_M_area):.0f})')
]):
    ax.fill(pts[0, :4], pts[1, :4], alpha=0.25, color=color)
    ax.plot(pts[0], pts[1], 'o-', color=color, lw=2, ms=7)
    ax.set_xlim(-0.5, 5); ax.set_ylim(-0.5, 3.5)
    ax.set_aspect('equal')
    ax.axhline(0, color='#ccc', lw=0.8); ax.axvline(0, color='#ccc', lw=0.8)
    ax.grid(True, linestyle='--', alpha=0.4)
    ax.set_title(title)
    ax.set_xlabel('x'); ax.set_ylabel('y')

plt.suptitle('|det(A)| = area scaling factor', fontsize=12)
plt.tight_layout()
plt.show()
```

## 7.5 Using the Determinant to Find Eigenvalues

The determinant is the key tool for finding **eigenvalues**. This section introduces the idea from scratch — no prior knowledge assumed.

### What Is an Eigenvector?

When you multiply a matrix $A$ by most vectors $\mathbf{v}$, both the direction and the length change. But certain special vectors — called **eigenvectors** — only get *scaled*; their direction stays the same:

$$A\mathbf{v} = \lambda\mathbf{v}$$

The scalar $\lambda$ is called the **eigenvalue** for that eigenvector. The word *eigen* is German for *own* — these are the matrix's own special directions.

**Example.** Let $A = \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}$ and $\mathbf{v} = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$.

$$A\mathbf{v} = \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}\begin{bmatrix} 1 \\ 0 \end{bmatrix} = \begin{bmatrix} 3 \\ 0 \end{bmatrix} = 3\begin{bmatrix} 1 \\ 0 \end{bmatrix}$$

So $\mathbf{v} = \begin{bmatrix}1\\0\end{bmatrix}$ is an eigenvector with eigenvalue $\lambda = 3$: multiplying by $A$ just stretches it by a factor of 3, leaving its direction unchanged.

### The Characteristic Equation

How do we *find* eigenvalues without guessing? Start from $A\mathbf{v} = \lambda\mathbf{v}$ and rearrange:

$$(A - \lambda I)\mathbf{v} = \mathbf{0}$$

For this to have a nonzero solution $\mathbf{v}$, the matrix $(A - \lambda I)$ must be **singular** — meaning its determinant equals zero:

$$\det(A - \lambda I) = 0$$

This is called the **characteristic equation**. The values of $\lambda$ that satisfy it are the eigenvalues.

### Finding the Eigenvalues of $A$

**Step 1.** Form $A - \lambda I$:

$$A - \lambda I = \begin{bmatrix} 3 - \lambda & 1 \\ 0 & 2 - \lambda \end{bmatrix}$$

**Step 2.** Apply the $2 \times 2$ determinant formula $ad - bc$:

$$\det(A - \lambda I) = (3 - \lambda)(2 - \lambda) - (1)(0) = \lambda^2 - 5\lambda + 6.$$

**Step 3.** Set equal to zero and factor:

$$\lambda^2 - 5\lambda + 6 = (\lambda - 2)(\lambda - 3) = 0 \implies \lambda = 2 \text{ or } \lambda = 3.$$



```python
A_eig = np.array([[3., 1.],
                  [0., 2.]])

# Verify each eigenvalue: det(A - lambda*I) should be 0
for lam in [2, 3]:
    d = np.linalg.det(A_eig - lam * np.eye(2))
    print(f'det(A - {lam}*I) = {np.round(d, 4)}')
```

```python
# Run this cell without modification — plot the characteristic polynomial
lambdas = np.linspace(-0.5, 5.5, 300)
dets = [np.linalg.det(A_eig - lam * np.eye(2)) for lam in lambdas]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(lambdas, dets, color='steelblue', lw=2, label='det(A - lambda*I)')
ax.axhline(0, color='black', lw=0.8)
ax.axvline(2, color='tomato',  lw=1.5, linestyle='--', label='lambda = 2')
ax.axvline(3, color='seagreen', lw=1.5, linestyle='--', label='lambda = 3')
ax.scatter([2, 3], [0, 0], color=['tomato', 'seagreen'], zorder=5, s=80)
ax.set_xlabel('lambda')
ax.set_ylabel('det(A - lambda * I)')
ax.set_title('Characteristic polynomial — zeros are the eigenvalues')
ax.legend(fontsize=10)
ax.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

### Verifying with NumPy

`np.linalg.eig` computes both eigenvalues and eigenvectors. The eigenvectors are returned as columns of a matrix.

```python
eigenvalues, eigenvectors = np.linalg.eig(A_eig)

print('Eigenvalues:', np.round(eigenvalues, 4))
print()
print('Eigenvectors (each column is one eigenvector):')
print(np.round(eigenvectors, 4))
```
