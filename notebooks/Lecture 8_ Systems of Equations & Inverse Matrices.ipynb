## 8.1 Matrix-Vector Equations

**Definition**. A <u>matrix-vector equation</u> is an equation of the form

\begin{align}
A\vec{x} = \vec{b}.
\end{align}

Think of this as similar to $A\vec{x}_t = \vec{x}_{t+1}$ with a key difference: with Markov processes we start with $\vec{x}_t$ and $A$ then multiply them to find the (as of yet) unknown vector $\vec{x}_{t+1}$. If you're thinking from the perspective of "solving" for something, $\vec{x}_{t+1}$ is the thing we're solving for.

With $A\vec{x} = \vec{b}$ we have matrix $A$ and vector $\vec{b}$ and we're solving for vector $\vec{x}$. This, it turns out, is usually much more complicated than knowing $\vec{x}$ and solving for $\vec{b}$.

To fully understand how to do this we need get a deeper perspective on what "finding a solution $\vec{x}$" actually means.






**Key Concept**. We can view solving $A\vec{x} = \vec{b}$ for $x$ as "find the weights to multiply the columns of $A$ so that the linear combination of these weights times the columns produces $\vec{b}$ as a result."

Suppose that $A$ is $m \times n$, and $A = [\vec{a}_1 | \vec{a}_2 | ... | \vec{a}_n]$ where each $\vec{a}_i$ represent the columns of $A$ and $\vec{x} = <x_1, x_2, ... x_n>$. Then $A\vec{x} = \vec{b}$ is equivalent to

\begin{align}
x_1\vec{a}_1 + x_2\vec{a}_2 + ... + x_n\vec{a}_n = \vec{b}.
\end{align}

**Question**. What are the dimensions of $\vec{b}$?

### Column Space.

If we consider the columns of a matrix to be individual vectors $\vec{a}_1$, $\vec{a}_2$, ..., $\vec{a}_n$, then we can define the <u>column space</u> of a matrix $A$ as

\begin{align}
C(A) & = \{x_1\vec{a}_1 + x_2\vec{a}_2 + ... + x_n\vec{a}_n : x_i \in \mathbb{R}\}.
\end{align}

"For a given vector $\vec{b}$, do there exist weights $x_1, ..., x_n$ such that I can get $\vec{b}$ via linear combinations of the columns of matrix $A$?" If yes, then $\vec{b}$ is in the column space of $A$.

### Solutions to Matrix-Vector Equations.

A solution $\vec{x}$ to the "system" $A\vec{x} = \vec{b}$ will fall into one of the three following cases:

1. There are no $x_i$ values that solve the system. It is impossible to get vector $\vec{b}$ from a linear combination of the columns of $A$. This is the "no solution" case.
2. There are only one set of $x_i$ values (equivalently, a unique vector $\vec{x}$) that solve the system. This is called the "one solution" or "unique solution" case.
2. There are many different combinations of $x_i$ values that will make linear combinations that produce vector $\vec{b}$. This is the "infinite number of solutions" case.

### Matrix Rank.

A set of vectors is <u>linearly independent</u> if no linear combination of the vectors in the set can produce any other vector in the set.

For any matrix A:

$rank(A) =$

*   the number of linearly independent columns
*   the "dimensionality" of the column space
*   number of columns containing "information."

## 8.2 Matrix Inverses

An $n \times n$ <u>identity matrix</u> has 1s along its diagonal and 0s elsewhere:

\begin{align} I = \begin{bmatrix}
    1 & 0 & 0 & \dots & 0 \\
    0 & 1 & 0 & \dots & 0 \\
    0 & 0 & 1 & \dots & 0\\
    \vdots & \vdots & \vdots & \ddots & \\
    0 &  0  & 0 & \dots & 1
    \end{bmatrix} .
\end{align}

**Examples**. What is $I\vec{x}$ for any vector $\vec{x}$? What is $IA$ and $AI$ for any matrix $A$?

The <u>inverse matrix</u> for $A$ is denoted $A^{-1}$ and has the property:

\begin{align} AA^{-1} = A^{-1}A = I. \end{align}

#### Conditions for a matrix $A$ to have a "full" inverse $A^{-1}$:


* $A$ must have $m = n$.
* $A$ must be "full-rank", ie, $rank(A) = m=n$.
