## 1.1 Linear Algebra Applicaton: Data Science

A <u>model</u> is a mathematical description of the relationship between two or more different variables. This kind of modeling is common in statistics, machine learning, dynamical systems, regression, and classification.

<u>Data science</u> is the process of using data to formulate a model that gives us insight (reveals connections, makes predictions, etc) about an underlying phenomena that we wish to study. Modern data science involves tools from statistics, computer science, machine learning, and other areas of mathematics (depending on the specific application). Linear Algebra lies at the heart of data science and modeling applications, which is why we study it specifically in this course.

**Example**: We are building a model to predict amount of money a customer will spend at website y using the amount of money they spent websites x and z as predictor variables. Here is some example data:

x  | z  |  y
----|----|----
2 | 3 | 8
4 | 6 | 16
3 | 7 | 17
5 | 2 | 9
3 | 2 | 7

There are many ways that we could use this data to build a model to predict how much money a new customer (who is not in the data set) will spend at website y. One way we could build a model is to assume that the amount spent at website y is a linear function of the amounts a customer spent at websites x and z:

\begin{align}
\hat{y} = w_1x + w_2z.
\end{align}

We use the "hat" symbol to distinguish between a model's *prediction* $\hat{y}$ and the actual data values $y$. To finish "building" this model we need to use the data to "learn" the model "weights" $w_1$ and $w_2$.

One way to find the weights using the data is to set up the following system of equations:

\begin{array}{lcl} 2w_1  + 3w_2 & = & 8
\\ 4w_1  + 6w_2 & = & 16
\\ 3w_1 + 7w_2 & = & 17
\\ 5w_1 +  2w_2 & = & 9
\\ 3w_1 + 2w_2 & = & 7 \end{array}

If we can find values for $w_1$ and $w_2$ that satisfy all 5 of these equations then we plug these numbers into $\hat{y} = w_1x + w_2z$ and use the resulting equation to predict $y$ values that aren't in our original dataset (practitioners would call this "model deployment"). 

The mathematical process of actually finding $w_1$ and $w_2$, and whether these values exist in the first place, is complicated. We'll spend a considerable amount of time in this course developing techniques and intuition to solve problems like these.

## 1.2 Vectors

A <u>vector</u> is a mathematical object that has *magnitude* (length) and *direction*.

Vectors come in two *orientations*:

- A <u>column vector</u>: $\mathbf{x} = \begin{bmatrix} x_{1} \\ x_{2} \\ \vdots \\ x_{n} \end{bmatrix}$. This is a *general* column vector in *n dimensional space*. It's "general" because we are stating its *components* (also called "entries" or "elements" or "features") by naming them ($x_1$, $x_2$, ... $x_n$) without being specific about what these numbers are. 

- The *index* is the number that tells you which component is which: $x_1$ is the first, $x_n$ is the "nth." Oftentimes $i$ is used as a "dummy variable" that tracks the index: $x_i$ could refer to any of the components; when $i = 2$ we'd be referring to $x_2$. 

- A <u>row vector</u>: $\mathbf{x}^T = \begin{bmatrix} x_{1} & x_{2} & \cdots & x_{n} \end{bmatrix}$. $T$ is called the <u>transpose</u> operator; it converts a column vector $\mathbf{x}$ to row vector $\mathbf{x}^T$. Column vectors are the "default" orientation, but we'll deal with both types in this course. 

### Examples.

$$\mathbf{v} = \begin{bmatrix} 2 & 3 & 7 \end{bmatrix}, \qquad \mathbf{x} = \begin{bmatrix} 2 & 4 & 3 & 5 & 3 \end{bmatrix}^T, \qquad \mathbf{0} = \begin{bmatrix} 0 & 0 & 0 & 0\end{bmatrix}^T, \qquad \mathbf{1} = \begin{bmatrix} 1 & 1\end{bmatrix}$$

## 1.3 Matrices

**Definition.** A **matrix** is a rectangular array of numbers with $m$ rows and $n$ columns, 
written $A \in \mathbb{R}^{m \times n}$. We can think of a matrix in two ways:

**As a collection of column vectors:**

$$A = \begin{bmatrix} \mathbf{a}_1 & \mathbf{a}_2 & \cdots & \mathbf{a}_n \end{bmatrix}$$

where each $\mathbf{a}_j \in \mathbb{R}^m$ is a column vector.

**As a stack of row vectors:**

$$A = \begin{bmatrix} \mathbf{a}_1^T \\ \mathbf{a}_2^T \\ \vdots \\ \mathbf{a}_m^T \end{bmatrix}$$

where each $\mathbf{a}_i^T \in \mathbb{R}^n$ is a row vector.

Both perspectives are useful — we'll see that the column view connects naturally to 
datasets, where each column represents a measured variable (or *feature*), and each 
row represents a single observation.

### Example.

$$A = \begin{bmatrix} 2 & 3 & 8 \\ 4 & 6 & 16 \\ 3 & 7 & 17 \\ 5 & 2 & 9 \\ 3 & 2 & 7 \end{bmatrix}$$




## 1.4 Vectors and Matrices in Python: the `numpy` Library

Python is an *object-oriented* programming language, which means that getting stuff done is all about creating (and sometimes naming) objects in the computer's working memory. Here are some object types that you can create with "base" Python:

| Type | Example | Description |
|------|---------|-------------|
| `int` | `x = 3` | Integer |
| `float` | `x = 3.0` | Decimal number |
| `str` | `x = "hello"` | Text |
| `bool` | `x = True` | True or False |
| `list` | `x = [1, 2, 3]` | Ordered, mutable sequence |
| `tuple` | `x = (1, 2, 3)` | Ordered, immutable sequence |
| `dict` | `x = {"a": 1}` | Key-value pairs |

Notice that vectors and matrices are not on that list. It's possible to represent them with lists and tuples, but a (much) better way is to use the `numpy` library. 

### Examples. 

Create some code cell(s) below and use the `numpy` library to create the vectors and matrices we saw the the examples above. 

````python
import matplotlib.pyplot as plt 
import numpy as np

# Define the vector
x = np.array([[-2], [3]]) # Note the placement of the brackets- this is what makes x a column vector in Python!

# Plot
fig, ax = plt.subplots()
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='steelblue')

# Formatting
ax.set_xlim(-3, 2)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal')
ax.set_title(r'Vector $\mathbf{x}$')

# Place axis labels just inside the plot limits
ax.text(1.6, -0.15, r'$x_1$', fontsize=12)
ax.text(0.1, 3.8, r'$x_2$', fontsize=12)

ax.set_xlabel('')
ax.set_ylabel('')

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Define the vector
x = np.array([1, 2]) # A column vector in Python. Note the position of the square brackets. 

# Plot
fig, ax = plt.subplots()
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='steelblue', label=r'$\vec{x} = [-1, 2]$')

# Formatting
ax.set_xlim(-3, 3)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black', linewidth=0.8)
ax.axvline(0, color='black', linewidth=0.8)
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend(fontsize=12)
ax.set_title(r'Vector $\vec{x}$')
ax.set_aspect('equal')

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([-1, 2, 3]) # This is a 1-d array (vector). Note the square brackets. 

fig, axes = plt.subplots(1, 3, figsize=(9, 3), subplot_kw={'projection': '3d'})

views = [(20, 45), (20, 135), (45, 225)]
titles = ['View 1', 'View 2', 'View 3']

for ax, (elev, azim), title in zip(axes, views, titles):
    ax.quiver(0, 0, 0, x[0], x[1], x[2], color='steelblue', label=r'$\vec{x} = (-1, 2, 3)$')
    ax.set_xlim([-3, 3])
    ax.set_ylim([-1, 4])
    ax.set_zlim([0, 4])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.view_init(elev=elev, azim=azim)
    ax.set_title(title)
    ax.legend(fontsize=10)

plt.suptitle(r'Vector $\vec{x} = (-1, 2, 3)$', fontsize=14)
plt.tight_layout()
plt.show()
````
