### Left-Inverse Matrices.

A <u>left-inverse</u> matrix $L$ for $A$ has the property that $LA = I$, but not necessarily $AL = I$ (it may or may not be a "right-inverse").

$A^{-1}$ is a <u>full inverse</u> matrix for $A$ because it is both a left and right-inverse: $A^{-1}A = AA^{-1} = I$.

Left-inverses are especially useful when $A$ is "tall," and $m >> n$. A useful formula for $L$ when $A$ is tall is $L = (A^TA)^{-1}A^T$. You can read about one-sided inverses in pages 136 - 138 of *Practical Linear Algebra for Data Science*.

**Exercise**. Convince yourself that $(A^TA)^{-1}A^T$ is a left-inverse for $A$. To do this, plug $(A^TA)^{-1}A^T$ into $LA$ and see what happens! Is $(A^TA)^{-1}A^T$ also a right-inverse for $A$?

Why do we care about a left-inverse? They can be useful in solving $A\vec{x} = \vec{b}$ when a full-inverse does not exist!
