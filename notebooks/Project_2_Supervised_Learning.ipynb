### Why exploration comes first

Choosing a dataset is not like choosing a homework problem where the answer is guaranteed to exist. Real data is unpredictable, and a dataset that sounds interesting can turn out to have no meaningful signal, a poorly defined response variable, or features that are too correlated to be useful. The professional habit is to explore several candidates before committing to one.

For each candidate dataset you consider, you should at minimum do the following in your scratch notebook:

- Load it and check its shape, column names, and data types.
- Identify a clear **response variable**: a numerical column you want to predict (regression) or a categorical column with two or more classes (classification). Does it have enough variation to make prediction interesting?
- Check the features. Are any nearly constant? Redundant? Missing for many rows?
- Make a scatter plot or two of features against the response. Does there appear to be a relationship, or does the response look independent of the predictors?
- For classification: check the class balance. If 99% of rows belong to one class, accuracy is a misleading metric and prediction is a different kind of challenge.
- For regression: run a quick single-feature least squares fit and compute $R^2$. Is there any signal at all?

None of this pre-work needs to be polished. The goal is just to get enough signal to know whether a dataset is worth committing to.

### What makes a dataset suitable

- At least 2 meaningful numerical feature columns to use as predictors.
- At least 100 rows.
- A clear response variable: a continuous numerical column for regression, or a categorical column with a small number of classes for classification.
- Real, meaningful features — not arbitrary ID numbers or raw timestamps.
- Some domain reason to believe the features predict the response. This could be prior knowledge, a visual hint from a scatter plot, or a reasonable hypothesis worth testing.
