## 2.1 Weighted Sums and the Dot Product

In Lecture 1, we built a simple prediction model:

$$y = w_1 x + w_2 z.$$

We chose $w_1 = 1$ and $w_2 = 2$ by hand so that the formula would perfectly reproduce the $y$ values in our first dataset — that was a **synthetic** dataset, meaning we constructed it so the model would work out exactly. In practice, real data never lines up that cleanly.

Now consider a variation of that dataset where some of the y values are slightly different:

| $\mathbf{x}$ | $\mathbf{z}$ | $\mathbf{y}$ |
|-----|-----|-----|
| 2   | 3   | 8   |
| 4   | 6   | 15  |
| 3   | 7   | 18  |
| 5   | 2   | 9   |
| 3   | 2   | 7   |

For reference, the individual entries are indexed as follows:

| $\mathbf{x}$ | $\mathbf{z}$ | $\mathbf{y}$ |
|-----|-----|-----|
| $x_1$ | $z_1$ | $y_1$ |
| $x_2$ | $z_2$ | $y_2$ |
| $x_3$ | $z_3$ | $y_3$ |
| $x_4$ | $z_4$ | $y_4$ |
| $x_5$ | $z_5$ | $y_5$ |

The individual $x$, $z$, and $y$ values still represent each customer's spending. But unlike the original dataset, no choice of $w_1$ and $w_2$ will make $y = w_1 x + w_2 z$ reproduce every $y$ value exactly. The best we can do is find weights that get as close as possible for this "imperfect" data — this is the **least-squares solution**, which we will study in detail later in this class. For now, we'll just state the result:

$$w_1 = 0.9128, \qquad w_2 = 2.0671, \qquad \mathbf{w} = \begin{bmatrix} 0.9128 \\ 2.0671 \end{bmatrix}.$$

Notice that these weights are close to $(1, 2)$, but not exactly. Because the model no longer reproduces each customer's $y$ value perfectly, we need to distinguish between:

- $y_i$ — the **actual** value for customer $i$
- $\hat{y}_i$ — the customer's **predicted** value generated by the model.

In the real world, models make predictions that are generally different from the actual data values, so we'll switch our notation to reflect this:

$$\hat{y} = w_1 x + w_2 z.$$

**The model's dot product formulation.**

For each customer $i$, define the row vector $\mathbf{c}_i = \begin{bmatrix} x_i & z_i \end{bmatrix}$. For each customer $i$ prediction is then:

$$\hat{y}_i = w_1 x_i + w_2 z_i = \mathbf{c}_i \cdot \mathbf{w}.$$

This is the key insight: **a weighted sum is a dot product.**

**Note on order.** Mathematically, the dot product is commutative — $\mathbf{c}_i \cdot \mathbf{w} = \mathbf{w} \cdot \mathbf{c}_i$ — so the scalar result is the same either way. Python, however, does care about the shapes of your vectors when it "multiplies" them via the dot product. So while $\mathbf{c}_i \cdot \mathbf{w}$ looks backwards, its actually more consistent with how Python does the dot product!

**Example.** For the first customer in the dataset:

$$\hat{y}_1 = \mathbf{c}_1 \cdot \mathbf{w} =  (2)(0.9128) + (3)(2.0671) = 1.8256 + 6.2013  = 8.0269.$$

The actual value is $y_1 = 8$, so the prediction is close but not exact.

## 2.2 Projections

The **projection** of $\mathbf{x}$ onto $\mathbf{y}$ answers the question: *how "much" of $\mathbf{x}$ points in the direction of $\mathbf{y}$?*

**Definition.** The projection of $\mathbf{x}$ onto $\mathbf{y}$ is the vector:

$$\text{proj}_{\mathbf{y}}\,\mathbf{x} = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{y}\|^2}\,\mathbf{y}$$

What is really happening here: the scalar $\dfrac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{y}\|^2}$ is multiplied by the vector $\mathbf{y}$ to "resize" it. This is **scalar multiplication** — a constant stretches or shrinks a vector while keeping its direction. The scalar is chosen so that the resulting vector's tip lands exactly on the shadow that $\mathbf{x}$ casts onto the line through $\mathbf{y}$.

## 2.3 Cosine Similarity

Recall from lecture that the dot product has a geometric interpretation:

$$\mathbf{x} \cdot \mathbf{y} = \|\mathbf{x}\|\,\|\mathbf{y}\|\cos(\theta).$$

Rearranging this equation gives the **cosine similarity** of $\mathbf{x}$ and $\mathbf{y}$:

$$\text{sim}(\mathbf{x}, \mathbf{y}) = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\|\,\|\mathbf{y}\|}$$

This is just $\cos(\theta)$, the cosine of the angle between the two vectors. Its range is $[-1, 1]$:

- $\text{sim} = 1$. The vectors point in exactly the same direction.
- $\text{sim} = 0$. The vectors are orthogonal.
- $\text{sim} = -1$. The vectors point in exactly opposite directions.

**Why does cosine similarity range between $-1$ and $1$?** Notice what the denominator is doing: dividing $\mathbf{x}$ by $\|\mathbf{x}\|$ and $\mathbf{y}$ by $\|\mathbf{y}\|$ produces the unit vectors $\hat{\mathbf{x}}$ and $\hat{\mathbf{y}}$. So cosine similarity is simply the dot product of the normalized vectors:

$$\text{sim}(\mathbf{x}, \mathbf{y}) = \hat{\mathbf{x}} \cdot \hat{\mathbf{y}}.$$

What's really happening here is that cosine similarity strips away magnitude entirely by "normalizing" the $\mathbf{x}$ and $\mathbf{y}$ vectors to be unit vectors $\hat{\mathbf{x}}$ and $\hat{\mathbf{y}}$, and in doing so compares only the vector's relative directions. Now think about the graph of $\cos(\theta)$: it goes up and down between $1$ and $-1$. This tells you that the largest $\text{sim}(\mathbf{x}, \mathbf{y})$ could ever be is $1$, and the smallest is could be is $-1$.

### Application: Word Embeddings and Cosine Similarity

 One of the most famous applications of cosine similarity is **Word2Vec** — a model developed by Google in 2013 that maps words to vectors in $\mathbb{R}^{300}$ such that words with similar meanings end up close together (high cosine similarity). The vectors are learned by training on a massive corpus of text: the core idea is that words appearing in similar contexts should have similar vector representations. For example, "king" and "queen" appear in similar contexts, so their vectors should be close together, and pointed in similar directions.

What makes these word embeddings in **Word2Vec** remarkable is that the vector arithmetic reflects semantic relationships. The most famous example is:

$$\mathbf{v}_{\text{king}} - \mathbf{v}_{\text{man}} + \mathbf{v}_{\text{woman}} \approx \mathbf{v}_{\text{queen}}.$$

In other words, the difference between "king" and "man" captures the concept of royalty stripped of gender — adding "woman" back recovers "queen". This kind of structure emerges automatically from training on text.

We'll use a pre-trained **GloVe** model (a variant of **Word2Vec** developed at Stanford) with 50-dimensional vectors — so each word is a vector in $\mathbb{R}^{50}$ instead of $\mathbb{R}^{300}$. There is, as always, a tradeoff happening here: using vectors in $\mathbb{R}^{300}$ allows for richer, more meaninful (and therefore accurate) embeddings than their lower dimensional counterparts in $\mathbb{R}^{50}$ at the expense of more memory and more computational power.

## 2.4 Metrics for Model Performance: RMS and RMSE

The **Root Mean Square (RMS)** of a vector $\mathbf{x} \in \mathbb{R}^n$ is:

$$\text{RMS}(\mathbf{x}) = \sqrt{\frac{1}{n}\sum_{i=1}^{n} x_i^2} = \sqrt{\frac{\mathbf{x} \cdot \mathbf{x}}{n}} = \frac{\|\mathbf{x}\|}{\sqrt{n}}.$$

The last form shows that RMS is simply the norm of $\mathbf{x}$ scaled down by $\sqrt{n}$. This normalization means RMS measures the **typical size** of an entry, while the norm measures the total size of the vector. A higher dimensional vector (larger $n$) isn't automatically "bigger" in the RMS sense — only if its entries are individually larger.

### Applying RMS to Model Errors: RMSE

Recall from Section 2.1 the error vector:

$$\boldsymbol{\epsilon} = \mathbf{y} - \hat{\mathbf{y}} = \begin{bmatrix} y_1 - \hat{y}_1 \\ y_2 - \hat{y}_2 \\ \vdots \\ y_5 - \hat{y}_5 \end{bmatrix}.$$

The **Root Mean Square Error (RMSE)** is simply RMS applied to this error vector:

$$\text{RMSE} = \text{RMS}(\boldsymbol{\epsilon}) = \frac{\|\boldsymbol{\epsilon}\|}{\sqrt{n}}.$$

This is widely used metric that measures the typical size of the model's prediction errors — one number that summarizes how well the $\hat{y} = w_1 x + w_2 z$ model "fits" (predicts) the data.

**Computing $\hat{\mathbf{y}}$ via matrix multiplication.** Rather than computing each $\hat{y}_i = \mathbf{c}_i \mathbf{w}$ separately as in Section 2.1, we can stack all five data vectors into a **design matrix** $C$ and compute all predictions at once:

$$\hat{\mathbf{y}} = C\mathbf{w}, \qquad C = \begin{bmatrix} \mathbf{c}_1 \\ \mathbf{c}_2 \\ \vdots \\ \mathbf{c}_5 \end{bmatrix} = \begin{bmatrix} 2 & 3 \\ 4 & 6 \\ 3 & 7 \\ 5 & 2 \\ 3 & 2 \end{bmatrix}.$$

This is actually doing the same thing as computing the errors individually, but using matrix algebra (matrix-vector multiplication, to be specific) to compute the $\hat{y}_i$ predictions all at once. We will cover this in detail later in the course. For now, just run the cell below after you complete the code to calculate `epsilon`.
